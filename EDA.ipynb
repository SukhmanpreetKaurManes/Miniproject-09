```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/winequality-red.csv',delimiter=';')   #delimiter (separator) between values is semicolon
print("Fist 5 rows of the dataset")                            # delimiter tells ki jo sadi file ohde vich apa separator lai ki use krea
df.head()
```

```python
df1=pd.read_csv('/content/winequality-white.csv',delimiter=';')
print("Last 5 rows of the dataset")
df1.tail()
```

```python
print("Column Information :")
df.info()
```

```python
print("Shape of dataset : ",df.shape)
print("\nColumns names :")
print(df.columns)                             # it is an attribute that is why () are not used
print("\n")
print("Keys names :")
print(df.keys())                              # both give names of labels in dataframes
```

```python
print("Summary statistics  : ")
print(df.describe())
```

```python
pd.unique(df['quality'])
```

```python
df=df.drop(["alcohol","chlorides"],axis=1)             # dropping irrelevent columns
df.head()
```

```python
#renaming columns
df=df.rename(columns={'fixed acidity':'fixed_acidity','volatile acidity':'volatile_acidity'})
df.head()
```

```python
duplicate_rows=df[df.duplicated()]
print("Number of duplicate rows : ",len(duplicate_rows))
print(duplicate_rows.head(5))
df.drop_duplicates(inplace=True)
duplicate_rows=df[df.duplicated()]
print("Number of rows after deletion of duplicates",len(duplicate_rows))
print("New shape of dataset",df.shape)
print("Dataset after removing duplicates : ")
df.head()
```

```python
df.count()
```

```python
# dropping the missing or null values
print(df.isnull().sum())
```

```python
df=df.dropna()
df.count()
```

```python
print(df.isnull().sum())   # after dropping
```

```python
sns.boxplot(x=df['quality'])
```

```python
plt.figure(figsize=(10,8))
sns.histplot(df['pH'])
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['density'],y=df['quality'])
plt.title("Scatter plot between density and quality")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=df['pH'],y=df['quality'])
plt.title("Line plot between pH and quality")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm',fmt='.2f')
plt.title("Correlation heatmap")
plt.show()
```
